```python
# Replace unknown/error values

df['Item']=df['Item'].replace(['UNKNOWN','ERROR'], pd.NA)
df['Quantity']=df['Quantity'].replace(['UNKNOWN','ERROR'], pd.NA)
df['Price Per Unit']=df['Price Per Unit'].replace(['UNKNOWN','ERROR'], pd.NA)
df['Payment Method']=df['Payment Method'].replace(['UNKNOWN','ERROR'], pd.NA)
df['Location']=df['Location'].replace(['UNKNOWN','ERROR'], pd.NA)
```

```python
# Check data type of each column
df.info()

# Change column data type

df['Quantity'] = df['Quantity'].astype('Int64')
df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'], errors='coerce')
df['Total Spent'] = pd.to_numeric(df['Total Spent'], errors='coerce')
df['Transaction Date']=pd.to_datetime(df['Transaction Date'], errors ='coerce')
```

```python
df.head(50)
```

```python
# Convert all <NA> to NaN

df = df.fillna(value=np.nan)
```

```python
# Renaming Items based on the price

#df.loc[(df['Price Per Unit'] == 1.5) & (df['Item'].isna()), 'Item'] = 'Tea'
#df.loc[(df['Price Per Unit'] == 2) & (df['Item'].isna()), 'Item'] = 'Coffee'
#df.loc[(df['Price Per Unit'] == 5) & (df['Item'].isna()), 'Item'] = 'Salad'
#df.loc[(df['Price Per Unit'] == 1) & (df['Item'].isna()), 'Item'] = 'Cookie'
df.head(40)
```

```python
# Correct Total Spent column

mask = df['Total Spent'].isna() & df['Quantity'].notna()
df.loc[mask, 'Total Spent'] = df['Quantity'] * df['Price Per Unit']
```

```python
# Drop NaN, <NA>, NAT

df = df.dropna()
df = df.reset_index(drop=True)
```
